# Steps involved in creating a neural network #

1) Define the model
2) Compile the model
3) Fit the model


```python
# Import iris dataset #
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
Y = iris.target
# X = (x for x in X if Y!=2)
# Y = (y for y in Y if Y!=2)
# Convert to dataframes/ numpy arrays #
# iris.data.shape
# print(Y)


# Split the dataset into test and train datasets

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.60, random_state=0)
```

```python
### Logistic regression using DNN ###
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.utils import np_utils

import numpy
numpy.random.seed(7)

# Defining model #

model = Sequential()
model.add(Dense(1,input_dim=4,activation='sigmoid'))
# model.add(Dense(16,activation='sigmoid'))
# model.add(Dense(1,activation='sigmoid'))

# Compile the model #

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model #

model.fit(x_train, y_train, epochs=150, batch_size=10)
# model.summary()

# Precision #

# Confusion matrix #


# Evaluate the model #

scores = model.evaluate(x_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

```

```python
# Prediction #
prediction = model.predict(x_test)
prediction = (prediction>0.5)
# prediction.shape
# print(prediction)

# Creating the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, prediction)
print(cm)
# Remove the 3rd class #
# Loss function and response variable encoding(y_observed) #
```

```python
### SVM using DNN ###

# Defining model #

model = Sequential()
model.add(Dense(1,input_dim=4,activation='linear'))
# model.add(Dense(1,activation='linear'))

# Compile the model #

model.compile(loss='categorical_hinge', optimizer='adam', metrics=['accuracy'])

# Fit the model #

model.fit(X, Y, epochs=150, batch_size=10)

# Evaluate the model #

scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
