## Diabetes dataset

### Testing sklearn ridge_regression api

```python
from sklearn.linear_model import Ridge, LinearRegression

import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# url = "https://s3.ap-south-1.amazonaws.com/mlsquare-datasets/diabetes.csv"
# data = pd.read_csv(url, delimiter=",", header=None, index_col=False)
# # sc = StandardScaler()
# # data = sc.fit_transform(data)
# # data = pd.DataFrame(data)


# X = data.iloc[:,:-1] #Why? 
# Y = data.iloc[:,-1]

from sklearn.datasets import load_diabetes

data = load_diabetes()

X = data['data']

Y = data['target']

x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

```python
sklearn_model = Ridge()

sklearn_model.fit(x_train, y_train)
```

```python
sklearn_model.score(x_test, y_test)
```

```python
y_pred = sklearn_model.predict(x_train)
```

```python
pred_test = sklearn_model.predict(x_test)
```

```python
sklearn_model.get_loss
```

### Ridge regression as dnn in keras

```python
## DNN implementation of ridge-regression

from keras.models import Sequential
from keras.layers.core import Dense
from keras.regularizers import l2

model = Sequential()
model.add(Dense(1, input_dim=10, activation='linear',
             kernel_regularizer=l2(0.1)))
model.compile(optimizer='adam', loss='mse')
model.fit(x_train, y_train, epochs=500, batch_size=50)
```

```python
y_pred_keras = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred_keras)
```

```python
model.evaluate(x_test, y_test)
```

## Housing price dataset

### Sklearn test

```python
from sklearn.linear_model import Ridge, LinearRegression

import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# load dataset
dataframe = pd.read_csv("../data/housing.csv", delim_whitespace=True, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:13]
Y = dataset[:,13]

x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

```python
sklearn_model = Ridge()

sklearn_model.fit(x_train, y_train)
```

```python
sklearn_model.score(x_test, y_test)
```

```python
x_train.shape
```

### Keras test

```python
## DNN implementation of ridge-regression

from keras.models import Sequential
from keras.layers.core import Dense
from keras.regularizers import l2

model = Sequential()
model.add(Dense(1, input_dim=13, activation='linear',
             kernel_regularizer=l2(0.1)))
model.compile(optimizer='adam', loss='mse')
model.fit(x_train, y_train, epochs=1000, batch_size=50)
```

```python
y_pred_keras = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred_keras)
```

```python
model.summary()
```

```python
model.get_config()
```

### IMLY test

```python
from mlsquare.imly import dope
from ray import tune
model = Ridge()

m = dope(model)


params = {
    "epochs": tune.grid_search([1000])
}

m.fit(x_train, y_train, verbose=1, params=params)
```

```python
m.score(x_test, y_test)

# Loss value
```

```python
y_pred_imly = m.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred_imly)
```
